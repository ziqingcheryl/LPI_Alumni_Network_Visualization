```python
import numpy as np
import pandas as pd
```

```python
def check_if_still_student(df_line,column_names):
    if pd.isnull(df_line['degree_start_year']):
        return False
    else:
        past_year = 2022 - df_line['degree_start_year']
        if 'year'+str(int(past_year)) in column_names:
            if df_line['year'+str(int(past_year))] == True:
                return True
            else:
                return False
```

```python
df1= pd.read_csv('data\sheet_1 - PhD Exp.csv')
df1.head(2)
```

Fill type of people who are still students

```python
df1['Type'].value_counts()
```

```python
df1['degree_start_year'] = df1['phd_year']
df1['Student']=df1.apply(lambda x: check_if_still_student(x,df1.columns),axis=1)
```

```python
df1['Student'].value_counts()
```

```python
df1['to_fill'] = df1.apply(lambda x: True if x['Type']=='Other' and x['Student']==True else False, axis=1)
```

```python
df1['Type'] = df1.apply(lambda x: 'Student' if x['to_fill']==True else x['Type'],axis=1)
df1['Type'].value_counts()
```

```python
df1_use = df1[['Name','Occupation','Country','Company','Position','Type']]
df1_use['CRI']='PhD'

df2_use = df2[['Name','Occupation','Country','Company','Position','Type']]
df2_use['CRI']='Licence Exp'

df3_use = df3[['Name','Occupation','Country','Company','Position','Type']]
df3_use['CRI']='Licence Exp'

df4_use = df4[['Name','Occupation','Country','Company','Position','Type']]
df4_use['CRI']='Master Exp'

df5_use = df5[['Name','Occupation','Country','Company','Position','Type']]
df5_use['CRI']='Master Exp'

df6_use = df6[['Name','Occupation','Country','Company','Position','Type']]
df6_use['CRI']='Licence Edu'

df7_use = df7[['Name','Occupation','Country','Company','Position','Type']]
df7_use['CRI']='Master Edu'

df8_use = df8[['Name','Occupation','Country','Company','Position','Type']]
df8_use['CRI']='Master Edu'
```

```python
df_use = pd.concat([df1_use,df2_use,df3_use,df4_use,df5_use,df6_use,df7_use,df8_use])
```

```python
df_use['Company'].value_counts()
```

```python
def concat_columns(column):
    column = column.dropna()
    return ','.join(set(column))
```

```python
df_use_new=df_use.groupby('Name').agg(concat_columns)
df_use_new
```
