```python
def convert_lists_to_sets(x):
    # print(x)
    final_list = []
    if len(x)==1 and type(x[0])==float:
        return set()
    elif len(x)==2 and type(x[0])==float and type(x[1])==float:
        return set()
    else:
        for i in x:
            final_list.extend(i)
        return set(final_list)
```

```python
df_use_new['Skills'] = df_use_new['Skills'].apply(lambda x: convert_lists_to_sets(x))
```

```python
df_use_new
```

```python
df_use_new['Name']=df_use_new.index
```

```python
df = df_use_new[['Name','Skills']]
```
